```python
class TiendaOnline:
    def __init__(self):
        # crear atributos
        self.inventario = []
        self.ventas_totales = 0.0  # 0.0 ya es float


# métodos para gestionar la TiendaOnline

    def agregar_producto(self, nombre, precio, cantidad):
        nombre = nombre.strip().casefold()  # normalizar entrada
        # buscar si existe
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad
                print(f"Información actualizada: {producto['nombre'].title()}, "
                      f"cantidad actual: {producto['cantidad']} unidades.")
                return
        # si no se encontró, se añade (fuera del for)
        self.inventario.append({
            'nombre': nombre,      # guardamos normalizado
            'precio': float(precio),  # convertimos a float
            'cantidad': cantidad
        })
        print(f"El producto {nombre.title()} con precio {precio} y cantidad {cantidad} "
              f"unidades, ha sido añadido al inventario.")


    def ver_inventario(self):
        for producto in self.inventario:
            print(f"Producto: {producto['nombre'].title()}, "
                  f"Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")


    def buscar_producto(self, nombre):
        nombre = nombre.strip().casefold()
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                print(f"Producto: {producto['nombre'].title()}, "
                      f"Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
                return
        print(f"El producto {nombre.title()} no se encuentra en el inventario.")


    def actualizar_stock(self, nombre, cantidad):
        nombre = nombre.strip().casefold()
        for producto in self.inventario:
            # normalizamos también el nombre almacenado
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad
                if producto['cantidad'] < 0:
                    producto['cantidad'] = 0
                print(f"Producto actualizado: {producto['nombre'].title()}, "
                      f"nueva cantidad: {producto['cantidad']} unidades.")
                return
        print(f"El producto {nombre.title()} no se encuentra en el inventario.")


    def eliminar_producto(self, nombre):
        nombre = nombre.strip().casefold()  # normalizar entrada
        # recorrer inventario (lista de diccionarios) por índices, para localizar clave 'nombre'
        for i in range(len(self.inventario)):
            # la i es el índice, range(len()) para obtener índices del inventario
            if self.inventario[i]['nombre'] == nombre:
                # comprobar el rango si el producto en la posición i tiene el nombre buscado
                self.inventario.pop(i)
                # pop(i) elimina el elemento en el índice i y “cierra” la lista
                print(f"El producto {nombre.title()} ha sido eliminado del inventario.")
                return
        print(f"El producto {nombre.title()} no se encuentra en el inventario.")


    def calcular_valor_inventario(self):
        valor_total = 0.0
        for producto in self.inventario:
            valor_total += producto['precio'] * producto['cantidad']
        print(f"Valor total del inventario: {valor_total} euros")
        return valor_total
```
